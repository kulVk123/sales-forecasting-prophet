```python
import pandas as pd
df = pd.read_csv('train.csv')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
sales_df = df.groupby('Date')['Weekly_Sales'].sum().reset_index()
sales_df.columns = ['ds', 'y']
sales_df.head()
```

```python
from prophet import Prophet
model = Prophet()
model.fit(sales_df)

future = model.make_future_dataframe(periods=90)
forecast = model.predict(future)
```

```python
from sklearn.metrics import mean_absolute_error
import numpy as np

actual = sales_df['y'][-90:].values
predicted = forecast['yhat'][:len(sales_df)][-90:].values

mae = mean_absolute_error(actual, predicted)
mape = np.mean(np.abs((actual - predicted) / actual)) * 100
print(f"MAE: {mae:.2f}, MAPE: {mape:.2f}%")
```

```python
model.plot(forecast)
```

```python
forecast[['ds','yhat']].to_csv('sales_forecast.csv', index=False)
```
